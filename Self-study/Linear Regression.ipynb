````python
# Importações:
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from random import randint
````

# Construção do Modelo Teórico

## Funções Utilizadas:

````python
# Função de Contrução de Valores:
def vetor_aleatorio(n):
  '''
    Função que constroi um Vetor pseudoaleatório de tamanho N.
  '''
  vetor = pd.Series([randint(0,100) for i in range(0,n)])
  return vetor

def vetor(n):
  '''
    Função que constroi
  '''
  vetor = pd.Series([float(input('Digite um valor: ')) for _ in range(n)])
  return vetor

def coef_ang(vetor_x, vetor_y):
  '''
    Função que calcula o coeficiente angular de uma
  regressão linear.
  '''

  # Checando se as entradas são validas conjuntamente:
  if len(vetor_x) != len(vetor_y):
    print("Erro!")
    return None

  # Calculando Elementos do Coeficiente Angular:
  n=len(vetor_x)
  somaproduto = sum(vetor_x * vetor_y)
  produtosoma = vetor_x.sum()*vetor_y.sum()
  somaquadrado=(vetor_x ** 2).sum()
  quadradosoma=(vetor_x.sum())**2

  # Calculando Coeficiente Angular:
  A1 = ((n*somaproduto)-produtosoma)/((n*somaquadrado)-quadradosoma)
  return A1

def Coef_lin(vetor_x,vetor_y,coef_angular=None):
  '''
    Função que calcula o coefieciente linear de uma
  regressão linear.
  '''

  # Checando se as entradas são validas conjuntamente:
  if len(vetor_x) != len(vetor_y):
    print("Erro!")
    return None

  xbarra = vetor_x.mean()
  ybarra = vetor_y.mean()

  # Caso já se queira passar o Coeficiente Angular:
  if coef_angular == int or coef_angular == int:
    A0 = ybarra-(vetor_x*coef_angular)
    return A0

  A0 = ybarra-(xbarra*coef_ang(vetor_x,vetor_y))
  return A0

def linear_function(x, vetor_x, vetor_y):
  '''
    Função que calcula a função linear em algum ponto
  '''

  A1 = coef_ang(vetor_x,vetor_y)
  A0 = Coef_lin(vetor_x,vetor_y)

  fp = A0+A1*x
  return fp

def regress_graf(vetor_x,vetor_y):
  # Criar pontos para a reta
  x_values = np.linspace(vetor_x.min(), vetor_x.max(), 100)
  y_values = linear_function(x_values, vetor_x, vetor_y)

  # Plotar o gráfico de dispersão
  plt.scatter(vetor_x, vetor_y, label='Pontos')

  # Plotar a reta de regressão
  plt.plot(x_values, y_values, color='#CD5C5C', label='Reta de Regressão')

  # Visualização:
  plt.xlabel('Valores de X')
  plt.ylabel('Valores de Y')
  plt.show()
````

## Teste:

````python
# Construção dos Vetores:
X=vetor_aleatorio(4)
Y=vetor_aleatorio(4)
````

````python
# Plotar o gráfico de dispersão
plt.scatter(X, Y, label='Pontos')

# Visualização:
plt.xlabel('Valores X')
plt.ylabel('Valores Y')
plt.show()
````

````python
# Calculando a Função Linear:
A1 = coef_ang(X,Y)
A0 = Coef_lin(X,Y)

# Função de Regressão Linear:
if A1 > 0:
  print(f'f(x) = {round(A0,2)}+{round(A1,2)}x')
elif A1 < 0:
  print(f'f(x) = {round(A0,2)}{round(A1,2)}x')

# Gráfico de regressão:
regress_graf(X,Y)
````
