```python
import numpy as np
from random import randint as rand
from matplotlib import pyplot as plt
```

```python
# Gerando um conjunto de dados lineares e calculando θ̂ usando o Método dos Mínimos Quadrados.
X = 2 * np.random.rand(100, 1)
Y = 4 + 3 * X + np.random.randn(100, 1)
```

# Teste sem padronizar:

```python
# Visualizando os dados
plt.scatter(X, Y, color='blue')
plt.xlabel('Variável Independente (x)')
plt.ylabel('Variável Dependente (y)')
plt.title('Dados para Regressão Linear Simples')
plt.grid(True)
plt.show()
```

```python
def reg(x,y):
  '''
    Função para gerar a regressão
  '''
  # Médias:
  xbarra = x.mean()
  ybarra = y.mean()

  # Coeficiente angular:
  b1 = np.sum((x - xbarra) * (y - ybarra)) / np.sum((x - xbarra) ** 2)

  # Intercept:
  b0 = ybarra - b1*xbarra

  # Valores previstos de y com base nos valores de x
  y_pred = b0 + b1 * x

  return y_pred, (b0,b1)
```

```python
y_pred, coefs = reg(X,Y)
```

```python
# Visualizando os dados e a reta de regressão
plt.scatter(X, Y, color='blue', label='Dados Observados')
plt.plot(X, y_pred, color='red', label='Reta de Regressão')
plt.xlabel('Variável Independente (x)')
plt.ylabel('Variável Dependente (y)')
plt.title('Dados e Reta de Regressão')
plt.legend()
plt.grid(True)
plt.show()
```

# Teste a variável preditora padronizada:

```python
xp = (X-X.mean())/np.std(X)
```

```python
y_prado_pred, coefs_padro = reg(xp,Y)
```

```python
# Visualizando os dados e a reta de regressão
plt.scatter(X, Y, color='blue', label='Dados Observados')
plt.plot(X, y_prado_pred, color='red', label='Reta de Regressão')
plt.xlabel('Variável Independente (x)')
plt.ylabel('Variável Dependente (y)')
plt.title('Dados e Reta de Regressão')
plt.legend()
plt.grid(True )
plt.show()
```

```python
# Analise padronização:
print(np.var(y_prado_pred))
print(np.std(y_prado_pred))
print(xp.mean())
print(coefs_padro)
```

```python
# Analise sem padronização:
print(np.var(y_pred))
print(np.std(y_pred))
print(X.mean())
print(coefs)
```

```python
def prec(v1, v2):
  '''
    Função construida para checar o quanto um
  valor se aproxima de outro.
  '''

  count = 0
  var1 = round(v1,count)
  var2 = round(v2,count)

  while var1 == var2:
    var1 = round(v1,count)
    var2 = round(v2,count)
    count += 1

  print(f'A diferença surgiu na {count}º casa decimal')
```

```python
prec(np.var(y_pred),np.var(y_prado_pred))
prec(np.std(y_pred),np.std(y_prado_pred))
```

```python
# Plotagem dos dados e da reta de regressão
plt.figure(figsize=(10, 5))

# Gráfico 1: Dados e Reta de Regressão
plt.subplot(1, 2, 1)
plt.scatter(X, Y, color='blue', label='Dados Observados')
plt.plot(X, y_pred, color='red', label='Reta de Regressão')
plt.xlabel('Variável Independente (x)')
plt.ylabel('Variável Dependente (y)')
plt.title('Dados e Reta de Regressão')
plt.legend()
plt.grid(True)

# Marcação do ponto de intercepto
plt.scatter(0,coefs[0], color='green', marker='o', label='Intercepto')
plt.legend()

# Gráfico 2: Exibição do valor de b1
plt.subplot(1, 2, 2)
plt.text(0.5, 0.5, 'Valor de β0: {:.2f} |'.format(coefs[0]), fontsize=12, ha='right')
plt.text(0.5, 0.5, '| Valor de β1: {:.2f}'.format(coefs[1]), fontsize=12, ha='left')
plt.axis('off')

plt.tight_layout()
plt.show()
```
