# Gerando dados artificiais:

```python
# importações:
import pandas as pd
import random
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
def cpf():
  '''
    Função geradora de números aleatórios, simulação
  de CPFs.

  Nota:
    Como eu quero gerar aleatóriamente número por número,
  do numero total, uma possivel solução é criar o numero
  como string e concatenar a string. Tendo em vista
  que não dá para simplismente concatenar uma lista de
  números.
  '''
  # Convertendo os elementos da lista em strings e juntando-os
  # Sequência de números de 0 a 8
  sequencia = list(range(9))

  # Embaralhar a sequência
  random.shuffle(sequencia)

  lista_de_strings = [str(x) for x in sequencia]
  numero_como_string = "".join(lista_de_strings)

  # Convertendo a string resultante em um número
  numero_cpf = int(numero_como_string)

  # Exibindo o número resultante
  return numero_cpf
```

```python
cpf()
```

```python
rank = ['bom','ruim']
```

```python
data = {'idade': [], 'conta_corrente': [], 'risco': []}

for n in range(1001):
    data['idade'].append(random.randint(18, 80))
    data['conta_corrente'].append(cpf())
    data['risco'].append(rank[random.randint(0, 1)])

df = pd.DataFrame(data)
df
```

```python
df['idade'].plot(kind='hist', bins=20, title='idade')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
df.groupby('risco').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
df['idade'].plot(kind='line', figsize=(8, 4), title='idade')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
figsize = (12, 1.2 * len(df['risco'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(df, x='idade', y='risco', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
# Definindo tamanho da visualização:
plt.figure(figsize=(15, 8))

# Definindo gráfico:
plt.scatter(df[df.risco == 'ruim'].idade, df[df.risco == 'ruim'].conta_corrente)
plt.scatter(df[df.risco == 'bom'].idade, df[df.risco == 'bom'].conta_corrente)

# Ajustando informações:
plt.xlabel('Idade')
plt.ylabel('CPF')
plt.legend(['ruim', 'bom'])
plt.show()
```
